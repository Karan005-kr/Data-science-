# From Python lists

```python
import pandas as pd

data = [['alice', 54], ['alpha', 74], ['ando', 34]]

df = pd.DataFrame(data, columns=["name","age"])
df
```

# from dictionary of lists

```python
import pandas as pd

data = {
    "Name" : ["slice" , "karan" , "komal"],
    "age" : ["23", "25", "24"]
}

df = pd.DataFrame(data)
df
```

# From numpy Arrays

```python
import numpy as ka

arr = ka.array([[1,2],[4,5]])
df = pd.DataFrame(arr, columns = ["A", "B"])
df
```
