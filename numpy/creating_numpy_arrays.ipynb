# Creating arrays from scratch:

```python
# for a matrix of all zero     use = zeros
import numpy as np
a = np.zeros((3, 3))
a
```

```python
# for a matrix all ones         use = ones

import numpy as np
b = np.ones((2, 4))
b
```

```python
# for any matrrix order but filled with any one elements

import numpy as np

c = np.full((2,2) , 8)
c
```

```python
# for any identity matrix of order n              use .eye
 
import numpy as np

d = np.eye(5)
d
```

```python
# here 1-1000 is range and 10 is the addition at every step

e = np.arange(1, 1000 , 10)
e
```

```python
# for equally spaced array

h = np.linspace(1, 100, 10)
h
```

# Changing Data Types


```python
# this is a explicit method


arr = np.array([[1,2,3],[4,5,6],[8,57,43],[12,21,543]], dtype='float32')
```

```python
arr
```

```python
# convert float into int

arr_int = arr.astype(np.int32)
arr_int
```
