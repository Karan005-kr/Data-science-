```python
import numpy as np

arr = np.array([1,2,3,4,5])
result = arr **2
result
```

# broadcasting

```python
arr = np.array([1,2,3,4,5])

result = arr + 10
result

# here 10 is broadcast across all element of an array
```

# Broad casting with Arrays of different Shapes


```python
# 2D array in 2D array

a1 = np.array([1,2,3])
a2 = np.array([4,5,6])

result = a1 + a2
result
```

```python
# 2D array in 1D array

a1 = np.array([[1,2,3], [9,8,7]])
a2 = np.array([4,5,6])

a1 + a2

```
