```python
# using python list


import numpy as np
import time

size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x + y for x , y in zip(l1, l2)]
end = time.time()
print(add[0:10])

print(end - start)
```

```python
# using numpy arrays


import numpy as np
import time

size = 1_000_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
end = time.time()

print(end - start)
```

- compare the time taken to multiply two python lists v/s two Numpy arrays

```python
import time

beta1 = list([1, 2, 3, 4, 6])
beta2 = list([7, 8, 9, 10, 11])
start = time.time()
multiply = list(x * y for x,y in zip(beta1 , beta2))
end = time.time()
print(multiply)
print("python list multiplication time:", end - start)
```

```python
import numpy as np
import time

b1 = np.array([1, 2, 3, 4, 6])
b2 = np.array([7, 8, 9, 10, 11])
start = time.time()
multiply = (b1 * b2)
end = time.time()
print(multiply)
print("python list multiplication time:", end - start)
```
